```python
import pandas as pd
import base64
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
```

```python
url = "http://bkref.com/pi/shareit/GPWx7"
html = pd.read_html(url, header=0)
team_per_game = html[0]
team_per_game
```

```python
team_per_game.drop([20, 31], axis=0, inplace=True)
team_per_game.drop(["Rk", "G", "MP"], axis=1, inplace=True)
team_per_game
```

```python
url = "http://bkref.com/pi/shareit/9l1iu"
html = pd.read_html(url, header=0)
team_opponent_per_game = html[0]
team_opponent_per_game
```

```python
team_opponent_per_game.drop([20, 31], axis=0, inplace=True)
team_opponent_per_game.drop(["Rk", "G", "MP"], axis=1, inplace=True)
team_opponent_per_game
```

```python
url = "http://bkref.com/pi/shareit/enmnS"
html = pd.read_html(url, header=1)
team_misc_stats = html[0]
team_misc_stats
```

```python
team_misc_stats.drop([15, 16, 32], axis=0, inplace=True)
team_misc_stats.drop(["Rk", "Age", "Arena", "Attend.", "Attend./G"], axis=1, inplace=True)
columns_to_rename = {"eFG%.1":"opponent_eFG%", "TOV%.1":"opponent_TOV%", "FT/FGA.1":"opponent_FT/FGA"}
team_misc_stats.rename(columns=columns_to_rename, inplace=True)
team_misc_stats
```

```python
all_team_stats = team_per_game.merge(team_opponent_per_game, on="Team", how="left", suffixes=("", "_opponent"))
all_team_stats
```

```python
all_team_stats = all_team_stats.merge(team_misc_stats, on="Team", how="left")
all_team_stats
```

```python
all_team_stats['Team'] = all_team_stats['Team'].str.replace('*', '')
all_team_stats
```

```python
url = "https://www.basketball-reference.com/leagues/NBA_2021_games-december.html#schedule"
html = pd.read_html(url, header=0)
december_schedule = html[0]
december_schedule
```

```python
url = "https://www.basketball-reference.com/leagues/NBA_2021_games-january.html#schedule"
html = pd.read_html(url, header=0)
january_schedule = html[0]
january_schedule
```

```python
url = "https://www.basketball-reference.com/leagues/NBA_2021_games-february.html#schedule"
html = pd.read_html(url, header=0)
february_schedule = html[0]
february_schedule
```

```python
url = "https://www.basketball-reference.com/leagues/NBA_2021_games-march.html#schedule"
html = pd.read_html(url, header=0)
march_schedule = html[0]
march_schedule
```

```python
url = "https://www.basketball-reference.com/leagues/NBA_2021_games-april.html#schedule"
html = pd.read_html(url, header=0)
april_schedule = html[0]
april_schedule
```

```python
url = "https://www.basketball-reference.com/leagues/NBA_2021_games-may.html#schedule"
html = pd.read_html(url, header=0)
may_schedule = html[0]
may_schedule
```

```python
full_schedule = pd.concat([december_schedule, january_schedule, february_schedule, march_schedule, april_schedule, may_schedule])
full_schedule
```

```python
columns_to_rename = {"PTS":"Away_PTS", "PTS.1":"Home_PTS"}
full_schedule.rename(columns=columns_to_rename, inplace=True)
full_schedule.drop(["Start (ET)", "Unnamed: 6", "Unnamed: 7", "Attend.", "Notes"], axis=1, inplace=True)
full_schedule
```

```python
full_schedule.dropna(subset=["Away_PTS", "Home_PTS"], inplace=True)
full_schedule
```

```python
full_schedule["Home/Away Winner"] = full_schedule.apply(lambda x: "Home" if x["Home_PTS"] > x["Away_PTS"] else "Visitor", axis=1)
full_schedule["Winning Team"] = full_schedule.apply(lambda x: x["Home/Neutral"] if x["Home_PTS"] > x["Away_PTS"] else x["Visitor/Neutral"], axis=1)
full_schedule["Margin of Victory"] = full_schedule.apply(lambda x: abs(x["Home_PTS"] - x["Away_PTS"]), axis=1)
full_schedule
```

```python
pwd = os.getcwd()
all_team_stats.to_csv(pwd + "/../datasets/team_year_stats.csv", index=False)
full_schedule.to_csv(pwd + "/../datasets/full_schedule.csv", index=False)
```
